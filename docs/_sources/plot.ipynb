```python
import torch, numpy as np, matplotlib.pyplot as plt
```

```python
checkpoint_files = [
    "f8c48.pth",
    "f16c192.pth",
]
for f in checkpoint_files:
    c = torch.load(f, map_location='cpu', weights_only=False)
    N = 501; dist = np.convolve(c['log_mse_losses'], N*[1/N], mode='valid')
    plt.plot(dist,label=f)
plt.legend()
```
